```python
## Checks the number of missing (null) values present in each column of the dataset

df.isnull().sum()
```

```python
# Separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Fill missing numerical values with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with mode
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

####  This code identifies numerical and categorical columns and fills missing values:
####  Numerical columns with median and categorical columns with mode.

```

```python
 for col in num_cols :
     Q1 = df[col].quantile(0.25)
     Q3 = df[col].quantile(0.75)
     IQR = Q3-Q1 
     outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR )]
     print(f"{col}: {outliers.shape[0]}outliers")
```

```python
df = df.drop_duplicates()
```

## Visualizing distribution of house prices


```python
##  This graph shows how house prices are distributed.
##  It helps identify skewness, common price ranges, and outliers.

plt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], kde=True,)
plt.title("Distribution of House Prices")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.show()
```

> **Observation:**  
> - Most houses are priced in the lower to mid range.  
> - The distribution is right‚Äëskewed with a few high‚Äëpriced houses.


# Visualising distribution of houses above ground living area 

```python
plt.figure(figsize=(8,5))
sns.histplot(df['GrLivArea'], kde=True)
plt.title('Distribution of Houses Above Ground Living Area')
plt.xlabel('Living Area (sq ft)')
plt.ylabel('Frequency')
plt.show()
```

> **Observation:**
> - Distribution is slightly right-skewed.
> - Most houses have moderate living area.
> - Very few large houses exist.


# Visualizing how many houses have different numbers of bedrooms


```python
plt.figure(figsize=(7,4))
sns.countplot(x='BedroomAbvGr', data=df, color='#4C72B0')
plt.title('Distribution of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

> **Observation:**  
> Most houses have 2 to 4 bedrooms, indicating standard residential layouts.


#  Outlier Identification 


```python
# Identifying outliers in SalePrice using IQR method

Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['SalePrice'] < lower_bound) | (df['SalePrice'] > upper_bound)]
outliers.shape
```

# BIVARIATE ANALYSIS

# Analyzing relationship between living area and sale price

```python
## This graph shows the relationship between house size and price.
## Larger living area generally results in higher house prices.

plt.figure(figsize=(7,4))
sns.scatterplot(x='GrLivArea', y='SalePrice', color='#EC7453', data=df)
plt.title("Living Area vs House Price")
plt.xlabel("Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.show()
```

> **Observation:**  
> - Sale price increases as living area increases.  
> - A few large houses have lower prices, indicating outliers.

# Analyzing impact of number of bedrooms on house prices

```python
## This boxplot shows how house prices vary with the number of bedrooms.
## It helps compare price distribution for houses having different bedroom counts.

plt.figure(figsize=(7,4))
sns.boxplot(x='BedroomAbvGr', y='SalePrice', color='#98C868', data=df)
plt.title("Bedrooms vs House Price")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Sale Price")
plt.show()
```

> **Observation:**  
> - Houses with 3‚Äì4 bedrooms have higher median prices.  
> - Higher bedroom counts show more price variation.

# Analysing the impact of Garage Capacity on Sale Price

```python
plt.figure(figsize=(7,4))
sns.boxplot(x='GarageCars', y='SalePrice', data=df)
plt.title('Garage Capacity vs Sale Price')
plt.xlabel('Garage Capacity')
plt.ylabel('Sale Price')
plt.show()
```

> **Observation:**
> - Houses with more garage capacity have higher median prices
> - Houses without garages generally have lower prices

# MULTIVARIATE ANALYSIS

### Correlation Heatmap

The heatmap shows the relationship between different numerical features.
It helps identify which features strongly affect house prices.


```python
# Computing correlation between numerical features
# Visualizing correlation matrix using heatmap

numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

> - **Observation:**
> - Area, Bathrooms, Bedrooms show strong correlation with price
> - Helps identify most influential variables


# Visualisation of pairwise relationship and potential linear trends

```python
sns.pairplot(df[['SalePrice', 'GrLivArea', 'OverallQual', 'GarageCars']])
plt.show()
```

> - **Observation:**
> - Clear linear trend between SalePrice and GrLivAra
> - Strong upward trend between SalePrice and OverallQual
> - GarageCars shows moderate influence

```python
df.groupby(['BedroomAbvGr', 'GarageCars'])['SalePrice'].mean()
```

> - **Observation:**

```python
plt.figure(figsize=(7,5))
sns.scatterplot( x='GrLivArea',y='SalePrice',hue='BedroomAbvGr',data=df)
plt.title('Living Area vs Sale Price by Bedroom Count')
plt.show()
```

> **Observation:**

```python
important_features = [ 'SalePrice','OverallQual','GrLivArea','TotalBsmtSF','GarageCars']
sns.heatmap(df[important_features].corr(),annot=True,cmap='Blues',fmt='.2f')
plt.title('Correlation Heatmap of Important Features')
plt.show()
```

### Observation:
- SalePrice has strong positive correlation with OverallQual and GrLivArea.
- This indicates that better quality and larger living area increase house prices.


```python
amenity_features = ['SalePrice','GarageArea','TotalBsmtSF','Fireplaces']
sns.heatmap(df[amenity_features].corr(),annot=True,cmap='Greens',fmt='.2f')
plt.title('Correlation Between Amenities and Sale Price')
plt.show()
```

### Observation:
- Garage area and basement size show moderate positive correlation with sale price.
- This suggests amenities play an important role in house valuation.
